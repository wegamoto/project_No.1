```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data with uneven sampling in x
x = [-3, -2, -1.6, -1.2, -.8, -.5, -.2, .1, .3, .5, .8, 1.1, 1.5, 1.9, 2.3, 3]
X, Y = np.meshgrid(x, np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# plot
fig, ax = plt.subplots()

ax.pcolormesh(X, Y, Z, vmin=-0.5, vmax=1.0)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the chart
categories = ['A', 'B', 'C', 'D']
group_1 = [20, 34, 30, 35]
group_2 = [25, 32, 34, 20]

# Number of bars per category
x = np.arange(len(categories))

# Bar width
width = 0.35

# Create the plot
fig, ax = plt.subplots()

# Create the bars for each group
bars1 = ax.bar(x - width/2, group_1, width, label='Group 1', color='skyblue')
bars2 = ax.bar(x + width/2, group_2, width, label='Group 2', color='orange')

# Add labels, title, and custom ticks for x and y axes
ax.set_xlabel('Categories')
ax.set_ylabel('Values (Units)')
ax.set_title('Grouped Bar Chart with Units')
ax.set_xticks(x)
ax.set_xticklabels(categories)

# Add a legend
ax.legend()

# Attach a text label above each bar in bars1 and bars2, displaying its height with units
def add_labels(bars, label):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height} {label}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1, 'units')
add_labels(bars2, 'units')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C'],
        'Value': [10, 20, 10, 15, 10, 12, 30, 25, 35]}

df = pd.DataFrame(data)

# Group by 'Category' and sum 'Value'
grouped_df = df.groupby('Category')['Value'].sum()

# Plot the grouped data
grouped_df.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Sum of Values')
plt.title('Sum of Values by Category')
plt.show()
```

```python
plt.title('Average ACT Scores by Region and Year')
plt.xlabel('Year')
plt.ylabel('Average ACT Score')
plt.xticks(grouped_df['year'].unique())
plt.legend(title='Region')
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
data = np.random.rand(10, 12)  # 10 rows, 12 columns
columns = [f'Feature {i}' for i in range(1, 13)]
index = [f'Sample {i}' for i in range(1, 11)]

# Create a DataFrame
df = pd.DataFrame(data, index=index, columns=columns)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df, annot=True, cmap='viridis', cbar=True)
plt.title('Sample Heatmap')
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
data = np.random.rand(10, 12)  # 10 rows, 12 columns

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

# Add titles and labels
plt.title('Random Data Heatmap')
plt.xlabel('Columns')
plt.ylabel('Rows')

# Show the plot
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame creation
data = {
    'act_state': ['Active', 'Active', 'Inactive', 'Inactive', 'Active', 'Inactive'],
    'composite': ['Low', 'High', 'Low', 'High', 'Medium', 'Medium'],
    'value': [10, 20, 15, 25, 30, 5]  # Replace with your actual data
}
df = pd.DataFrame(data)

# Pivot the DataFrame to create a matrix format for the heatmap
heatmap_data = df.pivot_table(index='act_state', columns='composite', values='value', aggfunc='sum')

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Value'})
plt.title('Heatmap of Act State vs Composite')
plt.xlabel('Composite')
plt.ylabel('Act State')
plt.show()
```
