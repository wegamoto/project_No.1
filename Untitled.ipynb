```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-axis values
y = [2, 3, 5, 7, 11]  # Y-axis values

# Create a plot
plt.plot(x, y, marker='o', color='b', linestyle='-')

# Add title and labels
plt.title('Sample Line Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show grid
plt.grid()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data with uneven sampling in x
x = [-3, -2, -1.6, -1.2, -.8, -.5, -.2, .1, .3, .5, .8, 1.1, 1.5, 1.9, 2.3, 3]
X, Y = np.meshgrid(x, np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# plot
fig, ax = plt.subplots()

ax.pcolormesh(X, Y, Z, vmin=-0.5, vmax=1.0)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the chart
categories = ['A', 'B', 'C', 'D']
group_1 = [20, 34, 30, 35]
group_2 = [25, 32, 34, 20]

# Number of bars per category
x = np.arange(len(categories))

# Bar width
width = 0.35

# Create the plot
fig, ax = plt.subplots()

# Create the bars for each group
bars1 = ax.bar(x - width/2, group_1, width, label='Group 1', color='skyblue')
bars2 = ax.bar(x + width/2, group_2, width, label='Group 2', color='orange')

# Add labels, title, and custom ticks for x and y axes
ax.set_xlabel('Categories')
ax.set_ylabel('Values (Units)')
ax.set_title('Grouped Bar Chart with Units')
ax.set_xticks(x)
ax.set_xticklabels(categories)

# Add a legend
ax.legend()

# Attach a text label above each bar in bars1 and bars2, displaying its height with units
def add_labels(bars, label):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height} {label}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1, 'units')
add_labels(bars2, 'units')

# Show the plot
plt.show()
```
